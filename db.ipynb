```python
# Create table for connect :

import pymysql
conn=pymysql.connect(user='root',password='Ravi@1998',host='localhost',
                             database='rest_review_db')
qur='create table reviews_table(food VARCHAR(40) PRIMARY KEY ,good_review INT,bad_review INT,customer INT)'
mycur=conn.cursor()
mycur.execute(qur)
mycur.close()
conn.close()

```

```python
# list of food's to reviews:

foods = ["Idly", "Dosa", "Vada", "Roti", "Meals", "Veg Biryani",
         "Egg Biryani", "Chicken Biryani", "Mutton Biryani",
         "Ice Cream", "Noodles", "Manchooriya", "Orange juice",
         "Apple Juice", "Pineapple juice", "Banana juice"]
```

```python
# Insert food names into table :

for i in foods:
    conn=pymysql.connect(user='root',password='Ravi@1998',host='localhost',
                         database='rest_review_db')
    qur='INSERT INTO reviews_table(food,good_review,bad_review,customer) VALUES("%s",%d,%d,%d)'%(i,0,0,0)
    mycur=conn.cursor()
    mycur.execute(qur)
    conn.commit()  # whenever we use data manupulation concept like upgrade or delete,insert then we use commit.
    mycur.close()
    conn.close()
```
